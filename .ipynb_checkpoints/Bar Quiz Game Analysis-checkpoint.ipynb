# Bar Quiz Game Analysis

This is going to be finding the best move that can be taken in the classic bar quiz game 'Higher or Lower'. The quizmaster takes 6 random cards from the pack, then shows you one card. You have to guess whether the next card is higher or lower than the card being shown. If the card is the same, you automatically lose.

What is interesting about this game is the fact that the cards are initially selected at random from the deck. You could be shown a jack and instinctively think to guess lower, but the rest of the cards could all be Queens, Kings and Aces.

```python
class game2():
    def __init__(self,deck):
        self.deck = deck
        self.r_choices = np.random.choice(range(52), 6, replace=False)
        #self.r_choices = [0,4,8,12,16,20]
        self.random_cards = [self.deck[x] for x in self.r_choices]
    def play(self):
        x = 0
        while x < 5:
            card = self.random_cards[x][0]
            next_card = self.random_cards[x+1][0]
            print(card)
            p_input = None
            if card < 8:
                p_input = 'h'
            elif card > 8:
                p_input = 'l'
            elif card == 8:
                roll = random.randint(0,1)
                if roll == 0:
                    p_input = 'l'
                else:
                    roll = 'h'
            if p_input == 'h' and next_card < card:
                game_result.append([0])
                n_of_turns.append([x+1])
                was_duplicate.append([0])
                print('You lose!')
                return ('You lose!')
            elif p_input == 'l' and next_card > card:
                print(next_card)
                game_result.append([0])
                n_of_turns.append([x+1])
                was_duplicate.append([0])
                print('You lose!')
                return ('You lose!')
            elif (p_input == 'h' or p_input == 'l') and card == next_card:
                print(next_card)
                game_result.append([0])
                n_of_turns.append([x+1])
                was_duplicate.append([1])
                print('You lose! Duplicate card!')
                return ('You lose! Duplicate card!')
            x += 1
        else:
            #print('butts')
            card = self.random_cards[4][0]
            next_card = self.random_cards[4+1][0]
            if p_input == 'h' and next_card > card:
                print(next_card)
                game_result.append([1])
                n_of_turns.append([x])
                was_duplicate.append([0])
                print('You win!')
                return ('You win!')
            elif p_input == 'l' and next_card < card:
                print(next_card)
                game_result.append([1])
                n_of_turns.append([x])
                was_duplicate.append([0])
                print('You win!')
                return ('You win!')
            elif (p_input == 'h' or p_input == 'l') and card == next_card:
                print(next_card)
                game_result.append([0])
                n_of_turns.append([x+1])
                was_duplicate.append([1])
                print('You lose! Duplicate card!')
                return ('You lose! Duplicate card!')
            else:
                print(next_card)
                game_result.append([0])
                n_of_turns.append([x+1])
                was_duplicate.append([0])
                print('You lose!')
                return ('You lose!')
            

    
    
```
