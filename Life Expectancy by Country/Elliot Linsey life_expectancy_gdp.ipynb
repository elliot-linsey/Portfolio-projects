Importing the modules that I will be using.

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

Loading the data into a DataFrame and examining it.

```python
df = pd.read_csv('all_data.csv')
print(df.head())
```

Renaming column "Life expectancy at birth (years)" to make it easier to implement within code.

```python
df.rename(columns={'Life expectancy at birth (years)': 'LEABY'}, inplace = True)
```

```python
print(df.head())
```

---

## Bar Charts To Compare Average

Plotting bar charts to have a higher level look at the average GDP of each of the countries. It is clear that the USA has the highest GDP of all the countries, whilst Zimbabwe is almost non-existent in comparison. The others vary with almost equal steps inbetween them.

```python
fig = plt.subplots(figsize=(8,8))
sns.barplot(data=df, x='Country', y='GDP')
plt.xticks(rotation=30)
plt.show()
```

Creating a bar chart that plots life expectancy rather than GDP. Here we can see that most of the countries have similar life expectancy, however Zimbabwe is still significantly below with an almost 30 year difference. This is interesting as it may show that having a high GDP may not affect the life expectancy, up to a point. 

```python
fig = plt.subplots(figsize=(8,8))
sns.barplot(data=df, x='Country', y='LEABY')
plt.xticks(rotation=30)
plt.show()
```

## Violin Plots To Compare Life Expectancy Distributions 

Using violin plots we can see in greater detail how the life expectancy is spread out through the countries. For example, we can see that Mexico has the greatest density of results based around the median, whilst China is skewed to the left with a high density at the median but a more elongated tapering off over time as well as a greater interquartile range. Zimbabwe has changed the most over the time period, with an extremely thin violin plot with the largest range of any of the countries, no one point appears to have a density peak, indicating that all the results are fairly evenly distributed. 

```python
fig = plt.subplots(figsize=(15,10)) 
sns.violinplot(data=df, x='Country', y='LEABY')
sns.set_style('whitegrid')
plt.show()
```

## Bar Plots Of GDP and Life Expectancy over time

These graphs are plotting the differences over time of both GDP and Life Expectancy. They give a greater indication of the rapidity of a countries growth. For example, you can see that China started by making small increases for the first 7 years, then its GDP begins to jump higher and higher easily overtaking Germany and by the end reaching up to where the USA started. 
You can also see that Germany started to rise quite consistently, then fell down for a year and plateaued until it began to climb again. This gives greater scope to more questioning than the previous bar graphs and we can try and correlate these results with world events. What prompted China to rapidly increase its GDP? What happened to Germany so that its development fell and halted for 2 years?  

```python
f, ax = plt.subplots(figsize=(10, 15)) 
ax = sns.barplot(data=df, x='Country', y='GDP', hue='Year')
plt.xticks(rotation=45)
plt.ylabel('GDP in Trillions of U.S. Dollars')
plt.show()
```

This is the same type of bar plot but using life expectancy rather than GDP. You can see that the climb is steadier for all the countries and more even, Mexico and the USA appear to be rising at the same gradient. The major difference is Zimbabwe, where the life expectancy actually dropped for the first 5 years before starting to rise again. Again this leads to more questioning of what was occurring in Zimbabwe to account for the drop, but also what occurred to account for the quite significant rise in life expectancy and what lessons can we learn from this? How best could we apply the changes we've seen in countries like Zimbabwe to other developing nations?  

```python
f, ax = plt.subplots(figsize=(10, 15)) 
ax = sns.barplot(data=df, x='Country', y='LEABY', hue='Year')
plt.xticks(rotation=45)
plt.ylabel('Life Expectancy at birth in Years')
plt.show()
```

## Line Plots

It is better to use line graphs to demonstrate changing variables over time. These are facetgrids showing the life expectancy and GDP changes over the data period. An issue I came across was normalising the data for the GDP values. Due to the large numbers involved and the vast differences between them, to get them to appear on the same graph meant that for countries with only minor changes, such as Zimbabwe and Chile, their lines appeared to be unmoving whilst China and the USA had clear line movement. Investigating how to provide a more intuitive scale is my next step.  

```python
g3 = sns.FacetGrid(df, col="Country", col_wrap=3, height=4)
g3 = (g3.map(plt.plot, "Year", "LEABY").add_legend())
plt.show()
```

```python
g3 = sns.FacetGrid(df, col="Country", col_wrap=3, height=4)
g3 = (g3.map(plt.plot, "Year", "GDP").add_legend())
plt.show()
```
